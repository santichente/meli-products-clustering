# Elección de modelo de embeddings de texto a Utilizar con K-means


## Modelo de Hugging Face multilingual-e5-large-instruct


Este script implementa el clustering de productos basado en embeddings generados por el modelo preentrenado de Hugging Face.

---

## Proceso

### 1. **Generación de Embeddings:**
   - Se utiliza el modelo preentrenado [`intfloat/multilingual-e5-large-instruct`](https://huggingface.co/intfloat/multilingual-e5-large-instruct) para generar representaciones vectoriales de los títulos (`title_std_lemma`).
   - Los embeddings se generan en GPU para acelerar el procesamiento.

### 2. **Determinación del Número Óptimo de Clusters:**
   - Se aplica el algoritmo de **KMeans** para diferentes valores de `k` (número de clusters).
   - Dos métricas principales se utilizan para evaluar la calidad del clustering:
     - **Inercia:** Mide la compacidad de los clusters. Se utiliza en el método del codo.
     - **Silhouette Score:** Evalúa la separación y cohesión de los clusters.

### 3. **Selección del Mejor `k`:**
   - El número óptimo de clusters (`k`) se selecciona basado en el puntaje máximo de Silhouette.

### 4. **Visualización de Resultados:**
   - **Gráfica del Método del Codo:** Muestra cómo varía la inercia con el número de clusters.
   - **Gráfica del Silhouette Score:** Ayuda a identificar el valor óptimo de `k` basado en la calidad del clustering.

---

## Ventajas

- **Representaciones Multilingües:** 
   - Los embeddings generados por [`intfloat/multilingual-e5-large-instruct`](https://huggingface.co/intfloat/multilingual-e5-large-instruct) son robustos y multilingües, lo que permite procesar títulos en diferentes idiomas.
   - Ideal para aplicaciones internacionales en marketplaces globales.
- **Optimización del Clustering:** 
   - La combinación del método del codo y el Silhouette Score garantiza la selección de un número de clusters adecuado.
- **Escalabilidad:**
   - El uso de GPU y modelos preentrenados permite escalar el análisis para grandes volúmenes de datos.

---


```python
from sentence_transformers import SentenceTransformer
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
import os
from tqdm import tqdm

# Configuración del entorno
os.environ["OMP_NUM_THREADS"] = "4"

# Cargar el modelo de embeddings en GPU
model = SentenceTransformer("intfloat/multilingual-e5-large-instruct").to("cuda")

# Generar embeddings para los títulos de productos
df["title_std_lemma"] = df["title_std_lemma"].astype(str)
title_embeddings = model.encode(
    df["title_std_lemma"].tolist(), convert_to_numpy=True, device="cuda"
)

# Determinar el número óptimo de clusters con el método del codo y métrica de Silhouette
inertia = []
silhouette_scores = []
K_range = list(range(2, 800, 20))

for k in tqdm(K_range, desc="Calculando KMeans para diferentes valores de k"):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=20)
    kmeans.fit(title_embeddings)
    inertia.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(title_embeddings, kmeans.labels_))

# Determinar el mejor k basado en la métrica de Silhouette
best_k_index = np.argmax(silhouette_scores)
best_k = K_range[best_k_index]
best_silhouette_score = silhouette_scores[best_k_index]

# Imprimir las métricas del mejor k
print("\n--- Resultados ---")
print(f"Mejor número de clusters (k): {best_k}")
print(f"Silhouette Score para k={best_k}: {best_silhouette_score:.4f}")
print(f"Inercia para k={best_k}: {inertia[best_k_index]:.4f}")


def plot_elbow_method(k_range, inertia_values, optimal_k, model_name):
    """
    Grafica el método del codo para determinar k, con detalle del mejor número de clusters.

    :param k_range: Rango de valores de k.
    :param inertia_values: Lista de valores de inercia para cada k.
    :param optimal_k: El mejor número de clusters basado en Silhouette Score.
    :param model_name: Nombre del modelo de embeddings utilizado.
    """
    plt.figure(figsize=(12, 6))
    plt.plot(k_range, inertia_values, marker="o", linestyle="-", color="steelblue", label="Inercia")
    plt.axvline(x=optimal_k, color="red", linestyle="--", label=f"Mejor k = {optimal_k}")
    plt.scatter(optimal_k, inertia_values[k_range.index(optimal_k)], color="darkred", s=100)

    plt.annotate(
        f"Mejor k = {optimal_k}",
        xy=(optimal_k, inertia_values[k_range.index(optimal_k)]),
        xytext=(optimal_k + 50, inertia_values[k_range.index(optimal_k)] + 10),
        arrowprops=dict(facecolor="red", arrowstyle="->", lw=1.5),
        fontsize=11,
        color="black",
    )

    plt.xlabel("Número de Clusters (k)", fontsize=12)
    plt.ylabel("Inercia", fontsize=12)
    plt.title(f"Método del Codo con {model_name} para determinar k", fontsize=14, pad=15)
    plt.grid(alpha=0.5)
    plt.legend(fontsize=10)
    plt.tight_layout()
    plt.show()


def plot_silhouette_scores(k_range, silhouette_values, optimal_k, optimal_score, model_name):
    """
    Grafica la métrica de Silhouette para evaluar clusters, con detalle del mejor número de clusters.

    :param k_range: Rango de valores de k.
    :param silhouette_values: Lista de valores de Silhouette Score para cada k.
    :param optimal_k: El mejor número de clusters basado en Silhouette Score.
    :param optimal_score: El valor máximo de Silhouette Score.
    :param model_name: Nombre del modelo de embeddings utilizado.
    """
    plt.figure(figsize=(12, 6))
    plt.plot(k_range, silhouette_values, marker="o", linestyle="-", color="steelblue", label="Silhouette Score")
    plt.axvline(x=optimal_k, color="red", linestyle="--", label=f"Mejor k = {optimal_k}")
    plt.scatter(optimal_k, optimal_score, color="darkred", s=100)

    plt.annotate(
        f"Mejor k = {optimal_k}\nSilhouette = {optimal_score:.4f}",
        xy=(optimal_k, optimal_score),
        xytext=(optimal_k + 50, optimal_score - 0.02),
        arrowprops=dict(facecolor="red", arrowstyle="->", lw=1.5),
        fontsize=11,
        color="black",
    )

    plt.xlabel("Número de Clusters (k)", fontsize=12)
    plt.ylabel("Silhouette Score", fontsize=12)
    plt.title(f"Métrica de Silhouette con {model_name} para evaluar clusters", fontsize=14, pad=15)
    plt.grid(alpha=0.5)
    plt.legend(fontsize=10)
    plt.tight_layout()
    plt.show()


# Nombre del modelo para incluir en las gráficas
model_name = "intfloat/multilingual-e5-large-instruct"

# Graficar el método del codo y la métrica de Silhouette
plot_elbow_method(K_range, inertia, best_k, model_name)
plot_silhouette_scores(K_range, silhouette_scores, best_k, best_silhouette_score, model_name)
```

## Modelo de Open AI "text-embedding-3-large"


Este script implementa el clustering de productos basado en embeddings generados por la API de OpenAI, analizando los títulos de los productos.

---

## Proceso

### 1. **Generación de Embeddings:**
   - Se utiliza el modelo [`text-embedding-3-large`](https://platform.openai.com/docs/guides/embeddings) de OpenAI para generar representaciones vectoriales de los títulos de los productos.
   - Los embeddings se generan para los títulos estandarizados y lematizados (`title_std_lemma`).

### 2. **Cálculo del Número Óptimo de Clusters:**
   - Se utiliza el algoritmo **KMeans** para diferentes valores de `k` (número de clusters).
   - Dos métricas clave se calculan:
     - **Inercia:** Mide la compacidad de los clusters y se visualiza en el gráfico del método del codo.
     - **Silhouette Score:** Evalúa la separación y cohesión de los clusters.

### 3. **Selección del Mejor Número de Clusters (`k`):**
   - El número óptimo de clusters se selecciona basado en el puntaje máximo del **Silhouette Score**.
   - Las métricas clave del mejor `k` se imprimen.

### 4. **Visualización de Resultados:**
   - **Método del Codo:** Muestra cómo varía la inercia con el número de clusters, resaltando el mejor `k`.
   - **Gráfico de Silhouette Score:** Ayuda a evaluar la calidad de los clusters en función de la métrica de Silhouette.

---

## Ventajas

- **Embeddings Potentes:** Utiliza un modelo avanzado de OpenAI que captura relaciones semánticas complejas en los títulos.
- **Optimización Automática:** Selecciona el número de clusters óptimo utilizando métricas confiables.
- **Visualización Intuitiva:** Proporciona gráficos claros y anotados que explican la selección del número de clusters.
- **Escalabilidad:** Puede manejar grandes volúmenes de datos gracias al uso de embeddings generados en la nube.

---

## Consideraciones

- **Requisitos:** La API de OpenAI debe estar configurada con una clave válida.
- **Costo:** El uso de la API de OpenAI podría implicar costos, dependiendo del volumen de textos procesados.
- **Velocidad:** El tiempo de procesamiento depende del tamaño del dataset y de la capacidad de la API.

---


```python
import os
import getpass
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
from tqdm import tqdm
from openai import OpenAI

# Configurar la clave API de OpenAI
def _set_env(var: str):
    """
    Establece una variable de entorno solicitando su valor al usuario si no existe.

    :param var: Nombre de la variable de entorno.
    """
    if not os.environ.get(var):
        os.environ[var] = getpass.getpass(f"{var}: ")


_set_env("OPENAI_API_KEY")
client = OpenAI()

# Función para obtener embeddings usando la API de OpenAI
def get_embeddings(text_list, model="text-embedding-3-large"):
    """
    Genera embeddings para una lista de textos usando la API de OpenAI.

    :param text_list: Lista de textos (strings) para generar embeddings.
    :param model: Modelo de embeddings de OpenAI a utilizar.
    :return: Numpy array con embeddings generados.
    """
    embeddings = []
    for text in tqdm(text_list, desc="Generando embeddings"):
        try:
            response = client.embeddings.create(model=model, input=text)
            embeddings.append(response.data[0].embedding)
        except Exception as e:
            print(f"Error procesando el texto: {text}\n{e}")
    return np.array(embeddings)


# Procesar los títulos de productos
df["title_std_lemma"] = df["title_std_lemma"].astype(str)
title_embeddings = get_embeddings(df["title_std_lemma"].tolist())

# Determinar el número óptimo de clusters con el método del codo y Silhouette
inertia = []
silhouette_scores = []
K_range = list(range(2, 800, 20))

for k in tqdm(K_range, desc="Calculando KMeans para diferentes valores de k"):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=20)
    kmeans.fit(title_embeddings)
    inertia.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(title_embeddings, kmeans.labels_))

# Determinar el mejor k basado en la métrica de Silhouette
best_k_index = np.argmax(silhouette_scores)
best_k = K_range[best_k_index]
best_silhouette_score = silhouette_scores[best_k_index]

# Imprimir las métricas del mejor k
print("\n--- Resultados ---")
print(f"Mejor número de clusters (k): {best_k}")
print(f"Silhouette Score para k={best_k}: {best_silhouette_score:.4f}")
print(f"Inercia para k={best_k}: {inertia[best_k_index]:.4f}")


# Graficar el método del codo con anotaciones
plt.figure(figsize=(10, 5))
plt.plot(K_range, inertia, marker="o", linestyle="-", color="steelblue", label="Inercia")
plt.axvline(x=best_k, color="red", linestyle="--", label=f"Mejor k = {best_k}")
plt.scatter(best_k, inertia[best_k_index], color="darkred", s=100)

# Anotación para el mejor k en el gráfico de codo
plt.annotate(
    f"Mejor k = {best_k}",
    xy=(best_k, inertia[best_k_index]),
    xytext=(best_k + 50, inertia[best_k_index] + 10),
    arrowprops=dict(facecolor="red", arrowstyle="->", lw=1.5),
    fontsize=11,
    color="black",
)
plt.xlabel("Número de Clusters (k)", fontsize=12)
plt.ylabel("Inercia", fontsize=12)
plt.title("Método del Codo con Embeddings para determinar k", fontsize=14, pad=15)
plt.grid(alpha=0.5)
plt.legend(fontsize=10)
plt.tight_layout()
plt.show()


# Graficar la métrica de Silhouette con anotaciones
plt.figure(figsize=(10, 5))
plt.plot(
    K_range, silhouette_scores, marker="o", linestyle="-", color="steelblue", label="Silhouette Score"
)
plt.axvline(x=best_k, color="red", linestyle="--", label=f"Mejor k = {best_k}")
plt.scatter(best_k, best_silhouette_score, color="darkred", s=100)

# Anotación para el mejor k en el gráfico de Silhouette
plt.annotate(
    f"Mejor k = {best_k}\nSilhouette = {best_silhouette_score:.4f}",
    xy=(best_k, best_silhouette_score),
    xytext=(best_k + 50, best_silhouette_score - 0.02),
    arrowprops=dict(facecolor="red", arrowstyle="->", lw=1.5),
    fontsize=11,
    color="black",
)
plt.xlabel("Número de Clusters (k)", fontsize=12)
plt.ylabel("Silhouette Score", fontsize=12)
plt.title("Métrica de Silhouette con Embeddings para evaluar clusters", fontsize=14, pad=15)
plt.grid(alpha=0.5)
plt.legend(fontsize=10)
plt.tight_layout()
plt.show()
```

### 1. **Configuración de la API de OpenAI:**

   - Se utiliza la biblioteca `openai` para interactuar con la API.

### 2. **Generación de Embeddings:**
   - Se utiliza el modelo `text-embedding-3-large` para generar embeddings.
   - Los textos de la columna `title_std_lemma` se procesan de forma iterativa.
   - En caso de error en un texto, se registra un valor `None` para mantener la consistencia del dataset.

### 3. **Almacenamiento de Resultados:**
   - Los embeddings generados se añaden como una nueva columna (`embeddings_title_std_lemma`) al DataFrame original.
   - El DataFrame actualizado se guarda en un archivo Pickle (`productos_con_embeddings.pkl`) para facilitar su reutilización.

---

## Ventajas

- **Embeddings Avanzados:** Utiliza un modelo robusto de OpenAI para generar representaciones vectoriales precisas.
- **Compatibilidad con Otros Modelos:** Los embeddings generados son útiles para tareas de clustering, clasificación y búsqueda semántica.
- **Persistencia de Datos:** Guardar los resultados en un archivo Pickle permite reutilizar los embeddings sin necesidad de volver a generarlos.

---

## Consideraciones

- **Costos:** El uso de la API de OpenAI puede generar costos según el volumen de datos procesados.
- **Velocidad:** El tiempo de procesamiento depende del tamaño del dataset y del tiempo de respuesta de la API.
- **Errores:** En caso de error en la generación de embeddings para un texto, se registra un valor nulo (`None`) para evitar interrupciones.


```python
import os
import pandas as pd
import numpy as np
from tqdm import tqdm
from getpass import getpass
from openai import OpenAI

# Configurar la clave de API como variable de entorno
def _set_env(var: str):
    if not os.environ.get(var):
        os.environ[var] = getpass(f"{var}: ")

_set_env("OPENAI_API_KEY")

# Inicializar el cliente de OpenAI
client = OpenAI()

# Función para obtener embeddings usando la API de OpenAI
def get_embeddings(text_list, model="text-embedding-3-large"):
    """
    Genera embeddings para una lista de textos usando la API de OpenAI.

    :param text_list: Lista de textos (strings) para generar embeddings.
    :param model: Modelo de embeddings de OpenAI a utilizar.
    :return: Lista de embeddings generados.
    """
    embeddings = []
    for text in tqdm(text_list, desc="Generando embeddings"):
        try:
            response = client.embeddings.create(
                model=model,
                input=text
            )
            embeddings.append(response.data[0].embedding)
        except Exception as e:
            print(f"Error procesando el texto: {text}\n{e}")
            embeddings.append(None)  # En caso de error, agregar None
    return embeddings

# Asegurarse de que la columna `title_std_lemma` esté como string
df['title_std_lemma'] = df['title_std_lemma'].astype(str)

# Generar los embeddings para la columna `title_std_lemma`
df['embeddings_title_std_lemma'] = get_embeddings(df['title_std_lemma'].tolist())

# Ver los primeros registros del DataFrame con los embeddings
print(df.head())

# Guardar el DataFrame con los embeddings en un archivo Pickle
df.to_pickle("productos_con_embeddings.pkl")
print("Embeddings del título guardados en 'productos_con_embeddings.pkl'")
```

## DBSCAN

Este script implementa el clustering de productos basado en embeddings generados previamente y utiliza el algoritmo **DBSCAN** para identificar grupos de productos similares. Se realiza una búsqueda de parámetros para encontrar la mejor configuración del modelo.

---

## Proceso

### 1. **Carga de Datos:**
   - El archivo `productos_con_embeddings.pkl` es cargado para acceder a los embeddings generados previamente para los títulos de los productos.
   - Los embeddings se extraen de la columna `embeddings_title_std_lemma` y se convierten en una matriz NumPy.

### 2. **Aplicación de DBSCAN:**
   - Se aplica el algoritmo **DBSCAN** con diferentes combinaciones de parámetros `eps` y `min_samples`.
   - **`eps`** define la distancia máxima entre dos muestras para considerarlas en el mismo cluster. Si el valor de `eps` es muy pequeño, es probable que más puntos se clasifiquen como ruido, mientras que un valor más grande puede incluir puntos que deberían estar en diferentes clusters.
   - **`min_samples`** es el número mínimo de puntos necesarios para formar un cluster. Un valor bajo puede resultar en la formación de más clusters pequeños, mientras que un valor alto podría llevar a la creación de menos clusters más grandes.

### 3. **Evaluación de la Calidad del Clustering:**
   - **Silhouette Score:** Evalúa la separación y cohesión de los clusters. Se calcula solo si existen más de un cluster válido, proporcionando una medida de cuán bien definidos están los clusters.
   - **Davies-Bouldin Score:** Mide la calidad del clustering evaluando la distancia entre los centros de los clusters y su dispersión. Un valor más bajo indica mejor separación entre los clusters.
   - **Noise Ratio:** Proporción de puntos clasificados como ruido (etiquetados como `-1`), es decir, aquellos que no pertenecen a ningún cluster.

### 4. **Búsqueda de Parámetros (Grid Search):**
   - Se realiza una búsqueda exhaustiva de los mejores parámetros (`eps` y `min_samples`) evaluando las combinaciones posibles.
   - La configuración que maximiza el **Silhouette Score** es seleccionada como la mejor configuración para el modelo. Esto asegura que los clusters formados sean lo más cohesivos y bien separados posibles.

### 5. **Visualización y Almacenamiento de Resultados:**
   - Los resultados de cada combinación de parámetros se almacenan en un DataFrame y se ordenan por el **Silhouette Score**.
   - Los resultados se guardan en un archivo CSV en el directorio `../data/results/` para su posterior análisis.

---

## Ventajas

- **Flexibilidad de DBSCAN:** DBSCAN es un algoritmo basado en densidad, lo que significa que no necesita conocer el número de clusters de antemano, siendo útil para datos con clusters de forma arbitraria.
- **Evaluación Integral:** Las métricas de **Silhouette Score** y **Davies-Bouldin Score** proporcionan una evaluación completa de la calidad de los clusters. Mientras que el Silhouette Score evalúa la cohesión y la separación, el Davies-Bouldin proporciona una visión general de la dispersión y la cercanía entre clusters.
- **Detección de Ruido:** El modelo es capaz de identificar puntos que no pertenecen a ningún cluster, etiquetándolos como ruido. Esto es útil cuando hay productos que no se ajustan a ninguna categoría definida.

---

## Consideraciones

- **Parámetros Sensibles:** Los parámetros de DBSCAN, como `eps` y `min_samples`, son sensibles al tipo de datos y deben ajustarse con cuidado. Elegir un valor adecuado para `eps` es crucial, ya que afecta directamente la cantidad de puntos clasificados como ruido.
- **Escalabilidad:** DBSCAN es más adecuado para datasets con menos ruido y bien definidos, y puede ser costoso computacionalmente en grandes datasets si los parámetros no son óptimos. En datasets grandes, se puede experimentar con diferentes combinaciones de parámetros para optimizar el rendimiento.

---

## Resultados

- **Silhouette Score:** Los parámetros que maximicen el **Silhouette Score** serán seleccionados como los mejores, lo que indica que los clusters son más cohesivos y mejor separados.
- **Proporción de Ruido:** Se reporta el porcentaje de productos que no son asignados a ningún cluster. Esto proporciona información sobre cuántos productos son considerados atípicos o irrelevantes para los grupos principales.

---

## Conclusión

DBSCAN es un algoritmo robusto y flexible que permite realizar clustering sin necesidad de especificar el número de clusters de antemano. Su capacidad para detectar ruido y manejar clusters de formas arbitrarias lo hace adecuado para este tipo de análisis de productos. La combinación de DBSCAN con la evaluación de la calidad del clustering mediante el **Silhouette Score** y **Davies-Bouldin Score** proporciona una herramienta poderosa para segmentar productos de manera efectiva.


```python
import pandas as pd
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score, davies_bouldin_score

# Cargar el DataFrame desde el archivo Pickle (embeddings ya generados)
df = pd.read_pickle("../data/embeddings/productos_con_embeddings.pkl")

# Extraer los embeddings de la columna 'embeddings_title_std_lemma'
X_emb = np.array(df['embeddings_title_std_lemma'].tolist())

# Función para aplicar DBSCAN
def dbscan_clustering(X_emb, eps, min_samples):
    """
    Aplica DBSCAN al conjunto de datos.
    :param X_emb: Matriz de características (embeddings).
    :param eps: Distancia máxima entre dos muestras para considerarlas en el mismo cluster.
    :param min_samples: Número mínimo de puntos necesarios para formar un cluster.
    :return: Etiquetas de clusters y el modelo entrenado.
    """
    dbscan = DBSCAN(eps=eps, min_samples=min_samples, metric='euclidean')
    labels = dbscan.fit_predict(X_emb)
    return labels, dbscan

# Rango de parámetros para el Grid Search
eps_values = [0.3, 0.4, 0.5, 0.7, 0.8]
min_samples_values = [2, 3, 4, 5]

# DataFrame para almacenar resultados
results = []

# Grid Search
for eps in eps_values:
    for min_samples in min_samples_values:
        # print(f"\nEvaluando DBSCAN con eps={eps} y min_samples={min_samples}")
        labels, dbscan = dbscan_clustering(X_emb, eps, min_samples)

        # Calcular porcentaje de ruido
        noise_ratio = (labels == -1).sum() / len(labels)
        
        # Inicializar métricas
        silhouette_avg = None
        davies_bouldin_avg = None
        
        # Calcular métricas si hay clusters válidos
        valid_clusters = labels[labels != -1]
        if len(set(valid_clusters)) > 1:  # Al menos dos clusters válidos
            silhouette_avg = silhouette_score(X_emb[labels != -1], valid_clusters)
            davies_bouldin_avg = davies_bouldin_score(X_emb[labels != -1], valid_clusters)

        # Guardar resultados
        results.append({
            "eps": eps,
            "min_samples": min_samples,
            "silhouette_score": silhouette_avg,
            "davies_bouldin_score": davies_bouldin_avg,
            "noise_ratio": noise_ratio
        })

# Convertir los resultados a DataFrame
results_df = pd.DataFrame(results)

# Mostrar resultados ordenados por Silhouette Score (descendente)
results_df = results_df.sort_values(by="silhouette_score", ascending=False)
print("\nMejores combinaciones de parámetros según Silhouette Score:")
display(results_df.head())


# Crear directorio para resultados si no existe
import os
results_dir = "../data/results"
os.makedirs(results_dir, exist_ok=True)

# Guardar results_df como CSV
results_path = os.path.join(results_dir, "dbscan_results_clustering_title.csv")
results_df.to_csv(results_path, index=False)
print(f"\nResultados guardados en: {results_path}")

```

## Análisis de Clustering con HDBSCAN

Este proceso realiza el clustering de productos utilizando el algoritmo **HDBSCAN** (Hierarchical Density-Based Spatial Clustering of Applications with Noise). A través de un **Grid Search** con diferentes configuraciones de parámetros, se evalúan los resultados mediante métricas como el **Silhouette Score** y **Davies-Bouldin Score**. Los resultados se guardan en un archivo CSV para su posterior análisis.

## ¿Qué es HDBSCAN?

**HDBSCAN** es un algoritmo de clustering basado en la densidad, que extiende el algoritmo DBSCAN para manejar clusters jerárquicos y proporcionar una clasificación más flexible, especialmente en casos de datos con diferentes formas y densidades. A diferencia de DBSCAN, HDBSCAN no requiere un número predefinido de clusters, ya que el algoritmo selecciona automáticamente los clusters más prominentes basados en la densidad de puntos.

### Ventajas de HDBSCAN:
- **Detección de ruido:** HDBSCAN maneja los puntos que no pueden ser asignados a ningún cluster (puntos de "ruido"), etiquetándolos como `-1`.
- **Clusters de diferentes formas:** El algoritmo puede identificar clusters con formas no esféricas, a diferencia de KMeans.
- **Jerarquía de clusters:** HDBSCAN puede trabajar con clusters jerárquicos, lo que significa que es más flexible y dinámico para estructurar los datos.

## Proceso

### 1. **Aplicación de HDBSCAN:**
   - **Parámetros**:
     - `min_cluster_size`: Tamaño mínimo de un cluster (se prueba con 2, 5, 10).
     - `min_samples`: Número mínimo de muestras en el vecindario de un punto para considerarlo en un cluster (se prueba con 2, 5).
     - `epsilon`: Umbral de densidad para la selección de clusters (se prueba con valores de 0.05 y 0.1).
   - **Ruido:** Los puntos que no cumplen con los requisitos de densidad son clasificados como ruido (`-1`).

### 2. **Evaluación del Clustering:**
   - **Silhouette Score:** Mide la calidad del clustering en cuanto a la cohesión y la separación de los clusters. Un valor cercano a 1 indica una buena separación.
   - **Davies-Bouldin Score:** Evalúa la separación entre clusters. Un valor bajo indica clusters bien separados.
   - **Porcentaje de Ruido:** Mide el porcentaje de puntos que no fueron asignados a ningún cluster.

### 3. **Grid Search:**
   - Se realiza un **Grid Search** para explorar diversas combinaciones de los parámetros `min_cluster_size`, `min_samples` y `epsilon` para encontrar la configuración que brinde los mejores resultados en términos de las métricas mencionadas.

### 4. **Almacenamiento de Resultados:**
   - Los resultados del análisis, incluidas las métricas de evaluación para cada combinación de parámetros, se almacenan en un archivo CSV para su posterior análisis.

## Consideraciones

- **Rango de parámetros:** Los parámetros de HDBSCAN deben ser ajustados según la naturaleza de los datos. El tamaño del cluster y el número de muestras afectan directamente los resultados.
- **Manejo de ruido:** El algoritmo es eficiente para manejar ruido en los datos, pero un alto porcentaje de ruido puede afectar la interpretación de los resultados.

```python
import numpy as np
from sklearn.cluster import HDBSCAN
from sklearn.metrics import silhouette_score, davies_bouldin_score
import pandas as pd

# Cargar el DataFrame desde el archivo Pickle (embeddings ya generados)
df = pd.read_pickle("../data/embeddings/productos_con_embeddings.pkl")

# Extraer los embeddings de la columna 'embeddings_title_std_lemma'
X_emb = np.array(df['embeddings_title_std_lemma'].tolist())

# Definir rangos de parámetros para Grid Search
min_cluster_sizes = [2, 5, 10]  # Tamaño mínimo de un cluster
min_samples_list = [2, 5]  # Número mínimo de muestras en el vecindario
epsilon_values = [0.05, 0.1]  # Epsilon para la selección del cluster

# Almacenar resultados
results = []

# Grid Search
for min_cluster_size in min_cluster_sizes:
    for min_samples in min_samples_list:
        for epsilon in epsilon_values:
            # print(f"\nEvaluando HDBSCAN con min_cluster_size={min_cluster_size}, min_samples={min_samples}, epsilon={epsilon}")
            
            # Aplicar HDBSCAN
            hdbscan_clusterer = HDBSCAN(
                min_cluster_size=min_cluster_size,
                min_samples=min_samples,
                metric='euclidean',
                cluster_selection_epsilon=epsilon
            )
            hdbscan_labels = hdbscan_clusterer.fit_predict(X_emb)
            
            # Calcular número de clusters válidos
            n_clusters = len(set(hdbscan_labels)) - (1 if -1 in hdbscan_labels else 0)
            
            # Calcular métricas solo para clusters válidos
            valid_points = hdbscan_labels != -1
            X_valid = X_emb[valid_points]
            labels_valid = hdbscan_labels[valid_points]
            
            silhouette_avg = None
            davies_bouldin = None
            
            if n_clusters > 1:  # Necesitamos al menos 2 clusters válidos
                silhouette_avg = silhouette_score(X_valid, labels_valid, metric='euclidean')
                davies_bouldin = davies_bouldin_score(X_valid, labels_valid)
            
            # Calcular porcentaje de ruido
            noise_ratio = (hdbscan_labels == -1).sum() / len(hdbscan_labels)
            
            # Guardar resultados
            results.append({
                "min_cluster_size": min_cluster_size,
                "min_samples": min_samples,
                "epsilon": epsilon,
                "n_clusters": n_clusters,
                "silhouette_score": silhouette_avg,
                "davies_bouldin_score": davies_bouldin,
                "noise_ratio": noise_ratio
            })

# Convertir resultados a DataFrame
results_df = pd.DataFrame(results)

# Ordenar por Silhouette Score en orden descendente
results_df = results_df.sort_values(by="silhouette_score", ascending=False)

# Mostrar las mejores configuraciones
print("\nMejores combinaciones de parámetros según Silhouette Score:")
display(results_df.head())

# Crear directorio para resultados si no existe
import os
results_dir = "../data/results"
os.makedirs(results_dir, exist_ok=True)

# Guardar results_df como CSV
results_path = os.path.join(results_dir, "hdbscan_results_clustering_title.csv")
results_df.to_csv(results_path, index=False)
print(f"\nResultados guardados en: {results_path}")


```

## Visualización de Clusters con Imágenes de Productos

Este proceso visualiza los productos pertenecientes a los clusters más grandes utilizando imágenes y títulos. Se utiliza la biblioteca **matplotlib** para crear subplots que muestran las imágenes de los productos junto con sus títulos.

### Proceso

1. **Identificación de Clusters Más Grandes:**
   - Se obtienen los 10 clusters con más productos utilizando `df['cluster'].value_counts().head(5)`.
   - Se imprime la cantidad de productos en cada uno de estos clusters.

2. **Configuración de Subplots:**
   - Para cada uno de los clusters más grandes, se configura una figura de subplots.
   - Se limita la visualización a un máximo de 10 productos por cluster para mantener la claridad.

3. **Descarga y Visualización de Imágenes:**
   - Para cada producto en el cluster, se descarga la imagen utilizando la URL almacenada en la columna `thumbnail`.
   - Se utiliza `requests` para obtener la imagen y `PIL` para abrirla.
   - Las imágenes se muestran en los subplots, y se añade el título del producto, limitado a dos líneas para evitar desbordamientos.

4. **Manejo de Errores:**
   - Si ocurre un error al descargar o mostrar una imagen, se muestra un mensaje de "Imagen no disponible".

5. **Ajuste de Diseño:**
   - Se ajusta el diseño de los subplots para asegurar que las imágenes y los títulos se muestren correctamente.
   - Se ocultan los subplots sobrantes si hay menos productos que subplots disponibles.

### Resultados

- **Visualización de Clusters:** Se muestran las imágenes y títulos de los productos para los clusters más grandes, permitiendo una inspección visual rápida de los productos agrupados.
- **Interactividad:** Aunque no es interactivo en el sentido de un gráfico de Plotly, la visualización permite una rápida comprensión de los productos en cada cluster.

### Consideraciones

- **Límites de Visualización:** Se limita a 10 productos por cluster para evitar sobrecargar la visualización.
- **Conexión a Internet:** Se requiere una conexión a Internet para descargar las imágenes de los productos.
- **Manejo de Errores:** Se implementa un manejo básico de errores para asegurar que el proceso continúe incluso si algunas imágenes no están disponibles.

---

Esta visualización es útil para obtener una visión rápida de los productos dentro de los clusters más grandes, facilitando la identificación de patrones o características comunes entre los productos agrupados.

```python
import matplotlib.pyplot as plt
from PIL import Image
import requests
from io import BytesIO
import numpy as np
from textwrap import wrap

# Obtener los 10 clusters más grandes
cluster_counts = df['cluster'].value_counts().head(5)
print("10 Clusters con más productos:")
print(cluster_counts)

# Configuración de la figura para subplots
for cluster_id in cluster_counts.index:
    cluster_data = df[df['cluster'] == cluster_id]
    num_products = min(len(cluster_data), 10)  # Limitar a 15 productos máximo
    print(f"\nCluster {cluster_id} - Mostrando {num_products} de {len(cluster_data)} productos:")

    # Determinar el número de filas y columnas dinámicamente
    num_cols = 5  # Máximo 5 columnas
    num_rows = (num_products + num_cols - 1) // num_cols  # Calcular filas necesarias
    fig, axes = plt.subplots(num_rows, num_cols, figsize=(num_cols * 3, num_rows * 4))
    fig.suptitle(f"Cluster {cluster_id} - Total: {len(cluster_data)} productos (mostrando {num_products})", fontsize=16)

    # Aplanar los ejes para iterar fácilmente
    axes = axes.flatten() if isinstance(axes, np.ndarray) else [axes]

    # Mostrar imágenes y títulos de los productos
    for idx, (index, row) in enumerate(cluster_data.head(num_products).iterrows()):
        try:
            # Descargar y cargar la imagen
            response = requests.get(row['thumbnail'], timeout=10)
            response.raise_for_status()
            image = Image.open(BytesIO(response.content))

            # Mostrar la imagen en el subplot
            axes[idx].imshow(image)
            axes[idx].axis('off')
            # Título del producto, limitado a dos líneas
            title = "\n".join(wrap(f"Título: {row['title_std']}", width=30))
            axes[idx].set_title(title, fontsize=8, wrap=True)
        except Exception as e:
            axes[idx].axis('off')
            axes[idx].set_title("Imagen no disponible", fontsize=8)
        
    # Ocultar subplots sobrantes si hay menos productos que subplots
    for ax in axes[num_products:]:
        ax.axis('off')

    # Ajustar diseño del subplot
    plt.tight_layout(rect=[0, 0, 1, 0.95])  # Ajustar espacio para el título
    plt.show()
```
